```python
import pandas as pd
import json
from sklearn.feature_extraction.text import CountVectorizer
import itertools
```

```python
data = []
with open('../hw-1-parser/output/rambler-news.json') as file:
    for line in file:
        data.append(json.loads(line))

data = pd.DataFrame(data)
```

```python
data.drop(['title', 'link'],inplace=True,axis=1)
data = data.iloc[:100,:]
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(data['text'])
```

```python
data['bow'] = [vectorizer.transform([i]).indices for i in data['text']]
```

```python
iverted_vocabulary = {v: k for k, v in vectorizer.vocabulary_.items()}
```

```python
idx = {}
```

```python
for key in iverted_vocabulary.keys():
    tmp = set()
    for i, bow in enumerate(data['bow']):
        if key in bow:
            tmp.add(i)
    idx[key] = tmp
```

```python
text = 'Путин выборы политика'
```

```python
def find(text):
    ids = vectorizer.transform([text]).indices
    if len(ids) < 1:
        return None
    else:
        for i in range(len(ids),0,-1):
            comb = [c for c in itertools.combinations(ids, i)]
            for k in comb:
                print('Words  - {}'.format([iverted_vocabulary[word] for word in k]))
                s = set.intersection(*[idx[l] for l in k])
                for _, i in enumerate(s):
                    print('Text #{}\n\n \t{}\n'.format(_,data.iloc[i]['text']))
```

```python
find(text)
```
